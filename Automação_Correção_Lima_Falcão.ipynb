```python
import re

# Expressões regulares
r_cpf = r"(\d{3}\.\d{3}\.\d{3}-\d{2})" 
r_autor = r"([A-Za-zÀ-ÿ]+(?: [A-Za-zÀ-ÿ]+)*)"
r_valor = r"Valor da causa\nR\$\s?(\d{1,3}(?:\.\d{3})*(?:,\d{2})?)"
r_comarca = r"Jurisdição\n([A-Za-zÀ-ÿ\s]+) - "
r_citacao = r"Autuação\n([^\n]+)\nÚltima distribuição"
r_vara = r"Órgão julgador\n([^\n]*?de)"

# Iterar sobre os processos no dicionário
for numero_processo, dados in dados_processos.items():
    autor_texto = dados['autor']
    vara_texto = dados['vara']

    # Dividir o texto em linhas para o autor e a vara
    autor_linhas = autor_texto.split('\n')
    vara_linhas = vara_texto.split('\n')

    # Encontrar os dados usando expressões regulares
    nome = re.match(r_autor, autor_linhas[1])  # Nome do autor na segunda linha
    cpf = re.search(r_cpf, autor_linhas[1])   # CPF na segunda linha
    advg = re.match(r_autor, autor_linhas[2]) # Nome do advogado na terceira linha
    valor_causa = re.search(r_valor, vara_texto)
    comarca = re.search(r_comarca, vara_texto)
    citacao = re.search(r_citacao, vara_texto)
    vara = re.search(r_vara, vara_texto)
    print(f"Processo: {numero_processo}")
    # Exibir os resultados
    if nome:
        print("Nome do autor:", nome.group(1))
    else:
        print("Nome do autor não encontrado")

    if advg:
        print("Nome do advogado:", advg.group(1))
    else:
        print("Nome do advogado não encontrado")

    if cpf:
        print("CPF encontrado:", cpf.group(1))
    else:
        print("CPF não encontrado")

    if comarca:
        print("Comarca:", comarca.group(1))
    else:
        print("Comarca não localizada")

    if valor_causa:
        print("Valor da causa:", valor_causa.group(1))
    else:
        print("Valor não encontrado")
    
    if citacao:
        print("Data da citação", citacao.group(1))
    else:
        print("Valor não encontrado")
    if vara:
        vara_corrigir = vara.group(1)
        vara_correta = vara_corrigir.replace(" de","")
        print("Vara:", vara_correta)
    else:
        print("Valor não encontrado")
    print("\n" + "-"*50 + "\n")
```
